De forma gráfica observaremos cómo rinden ambos modelos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#CARGAR CSV
df = pd.read_csv("C:/Users/calag/Desktop/nlp-proyecto6/benchmarks/results/resumen_benchmark.csv")
print(df)
```

```python
#GRÁFICA DE MÉTRICAS BLEU y ROUGE
plt.figure(figsize=(10, 5))
sns.barplot(data=df.melt(id_vars="Modelo", value_vars=["BLEU", "ROUGE-L"]), x="variable", y="value", hue="Modelo")
plt.title("Comparación de Calidad de Traducción")
plt.ylabel("Puntaje")
plt.xlabel("Métrica")
plt.show()
```

```python
#GRÁFICA DE LATENCIAS
plt.figure(figsize=(10, 5))
sns.barplot(data=df.melt(id_vars="Modelo", value_vars=["p50_ms", "p95_ms", "p99_ms"]), x="variable", y="value", hue="Modelo")
plt.title("Comparación de Latencias")
plt.ylabel("Milisegundos")
plt.xlabel("Percentil")
plt.show()
```

```python
# MEMORIA RAM y VRAM
plt.figure(figsize=(10, 5))
sns.barplot(data=df.melt(id_vars="Modelo", value_vars=["ram_mb", "vram_mb"]), x="variable", y="value", hue="Modelo")
plt.title("Consumo de Memoria")
plt.ylabel("MB")
plt.xlabel("Tipo de Memoria")
plt.show()
```
